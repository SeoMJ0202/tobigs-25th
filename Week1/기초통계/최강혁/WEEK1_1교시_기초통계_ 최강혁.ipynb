## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

```python
import math
from scipy.stats import norm

p1 = 0.42  #A안: 전체 인기 공모전 섹션 도달률(기존)
p2 = 0.50  #B안: 개선 후 도달률(가정)

alpha = 0.05
power = 0.80

Z_a2 = norm.ppf(1 - alpha/2)   # Z_(alpha/2)
Z_b  = norm.ppf(power)         # Z_beta where beta=1-power

n = ((Z_a2 + Z_b)**2 * (p1*(1-p1) + p2*(1-p2))) / ((p1 - p2)**2)

print("n per group =", math.ceil(n))
print("total users =", math.ceil(n) * 2)
```

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
import math
import numpy as np
from scipy.stats import chi2_contingency
from statsmodels.stats.proportion import proportions_ztest

# =========================
# 0) 가정/설정값
# =========================
alpha = 0.05

daily_visitors = 404   # 가정한 하루 평균 방문자 수
n_per_group = 606      # 산출된 최소 test users 수 (각 그룹당)

# =========================
# 1) 관측 데이터 (성공, 실패)
# =========================
# 성공=상세페이지 진입, 실패=미진입
control_results = [11500, 34500]   # A: 하단 노출 (Control)
variant_results = [12190, 33810]   # B: 최상단 노출 (Variant)

# 2x2 분할표: [ [A성공, A실패], [B성공, B실패] ]
table = np.array([control_results, variant_results])

# =========================
# 2) CVR 계산
# =========================
A_success, A_fail = control_results
B_success, B_fail = variant_results

A_n = A_success + A_fail
B_n = B_success + B_fail

pA = A_success / A_n
pB = B_success / B_n

print("=== 기본 요약 ===")
print(f"A(Ctrl) CVR = {pA:.6f} ({A_success}/{A_n})")
print(f"B(Var)  CVR = {pB:.6f} ({B_success}/{B_n})")
print(f"Lift (pB - pA) = {pB - pA:.6f} ({(pB/pA - 1)*100:.2f}%)")
print()

# =========================
# 3) 카이제곱 검정 (참고: 기본은 '차이 존재'를 보는 양측 성격)
# =========================
chi2, p_chi2, dof, expected = chi2_contingency(table, correction=False)

print("=== 카이제곱 검정(독립성) ===")
print(f"Chi2 = {chi2:.4f}")
print(f"P-value(chi-square) = {p_chi2:.10f}")

if p_chi2 < alpha:
    print(f"결론: p < {alpha} → (차이가 있음) 귀무가설 기각")
else:
    print(f"결론: p ≥ {alpha} → (차이 증거 부족) 귀무가설 기각 불가")
print()

# =========================
# 4) 네 가설(H1: pA < pB)에 딱 맞는 단측 검정: 두 비율 z-test
#    H0: pA = pB
#    H1: pA < pB  (즉, B가 더 크다)
# =========================
count = np.array([A_success, B_success])   # 성공 수
nobs  = np.array([A_n, B_n])               # 전체 수

z_stat, p_one_sided = proportions_ztest(count, nobs, alternative="smaller")

print("=== 두 비율 z-test (단측, H1: pA < pB) ===")
print(f"Z = {z_stat:.4f}")
print(f"P-value(one-sided) = {p_one_sided:.10f}")

if p_one_sided < alpha:
    print(f"결론: p < {alpha} → 귀무가설 기각")
    print("=> 최상단 노출(B)이 하단 노출(A)보다 상세페이지 진입률(CVR)을 유의미하게 높임")
else:
    print(f"결론: p ≥ {alpha} → 귀무가설 기각 불가")
    print("=> B가 A보다 높다고 통계적으로 말하기 어려움")
print()

# =========================
# 5) 기간 산출 (너가 적은 값과 정합되게)
# =========================
total_required_users = n_per_group * 2
required_days = math.ceil(total_required_users / daily_visitors)

print("=== 실험 기간 산출(가정 기반) ===")
print(f"가정한 하루 평균 방문자 수 = {daily_visitors}명")
print(f"산출된 최소 test users 수 = {n_per_group}명 (그룹당) / 총 {total_required_users}명")
print(f"그래서 필요한 최소 기간 = {required_days}일")
```
