## A1. 답안 작성  
답변: A/B 테스트는 임의화 대조군 실험의 일종으로, 주로 웹사이트나 앱 서비스에서 사용자 경험이나 비즈니스 성과를 최적화하기 위해 사용되는 데이터 분석 방법론이다.

전체 사용자를 무작위로 두 개(또는 그 이상)의 집단으로 나누고, 대조군에게는 기존 서비스를, 실험군에게는 새로운 기능을 노출한 뒤, 어떤 집단이 더 높은 성과(클릭률, 구매 전환율 등)를 보이는지 통계적으로 검증하는 과정이다.

예를 들어, 핀테크 앱 메인 화면에서 '송금하기' 버튼의 클릭률(CTR)을 높이고자 할때, 송금 버튼을 하단 탭바 중앙에 크게 배치하면, 기존 상단 배치보다 접근성이 좋아져 송금 시도 횟수가 증가할 것이라고 가정하고 대조군과 실험군에게 일일 활성 사용자 대비 송금 버튼 클릭 비율을 일정 기간 테스트한다. 결과적으로 실험군의 클릭률이 대조군보다 통계적으로 유의미하게 높다면 변경된 앱을 선택한다.

A/B 테스트 진행 시 유의할 사항으로는,

첫 번째로 초두 효과가 있다.
초두 효과는 실험 초기 단계에서 발생하는 편향을 의미하며, 신기효과와 변화 저항으로 나타날 수 있다.
신기효과는 사용자가 변경된 UI가 더 좋아서가 아니라, 단지 새롭기 때문에 호기심에 클릭해보는 현상이다. 이로 인해 초기에 성과가 일시적으로 급등했다가 시간이 지나면 원래대로 돌아간다.
반대로, 변화 저항은 익숙한 인터페이스가 바뀌었을 때 느끼는 낯설음 때문에 변경된 UI의 효율이 일시적으로 급락하는 경우이다.

두 번째로 표본 크기와 통계적 유의성이 있다.
단순히 "B안의 클릭률이 1% 더 높다"라고 해서 성공을 단정 지을 수 없다. 그 차이가 우연에 의한 것인지, 실제 효과인지 검증해야 한다. 사전에 검정력과 유의수준을 고려하여, 결과를 신뢰할 수 있는 최소 표본 크기를 계산하고 테스트를 시작해야 한다.

세 번째로 외부 요인 통제가 있다. 테스트 기간 동안 특정 이벤트가 발생하면 데이터가 오염될 수 있다. 대조군과 실험군은 반드시 동일한 기간에 실험이 진행되어야 하며, 주말과 평일의 사용자 패턴이 다르므로 최소 1주일 단위로 테스트 주기를 설정하는 것이 일반적이다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

*   p1 (기존 전환율): 현재 데모 사이트의 메인 홈에서 공모전 상세 페이지로 진입하는 비율(CVR)을 의미한다. 기존 방문자 대비 Submit 비율인 15% (0.15)로 가정했다.
*   p2 (목표 전환율): 개선된 UI를 적용했을 때 기대하는 최소한의 전환율이다. UI 개선을 통해 17% (0.17)까지 2%p 상승시키는 것을 목표로 가정했다.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.

import math
from scipy.stats import norm

# 1. 설정값 입력
alpha = 0.05  # 유의수준 (Type I error)
beta = 0.2    # 1 - 검정력 (Type II error, Power=0.8)
p1 = 0.15     # 대조군(A) 예상 전환율 (15%)
p2 = 0.17     # 실험군(B) 예상 전환율 (17%)

# 2. Z값 구하기
z_alpha = norm.ppf(1 - alpha / 2) # 양측 검정
z_beta = norm.ppf(1 - beta)

# 3. 최소 표본 수(n) 계산 공식 적용
# n = (Z_alpha/2 + Z_beta)^2 * [p1(1-p1) + p2(1-p2)] / (p1 - p2)^2
numerator = (z_alpha + z_beta)**2 * (p1 * (1 - p1) + p2 * (1 - p2))
denominator = (p1 - p2)**2
n = numerator / denominator

print(f"그룹당 필요한 최소 표본 수 (n): {math.ceil(n)}명")
print(f"총 필요한 최소 표본 수 (Total): {math.ceil(n) * 2}명")

#########
```

[실험 가설]  
귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$


<br>

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱 검정 이고,  

이유는 비교하고자 하는 데이터가 '상세 페이지 진입 성공/실패'라는 범주형 변수이며, 두 집단 간의 '비율' 차이가 유의미한지 검증해야 하기 때문이다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (Q2에서 설계한 표본 수와 전환율을 바탕으로 가정)
# [전환 성공(Submit), 전환 실패]
# 대조군(A): 약 5,271명 중 15% 전환 가정 -> 약 791명 성공
control_results = [791, 4480]

# 실험군(B): 약 5,271명 중 17.5% 전환 가정 (목표치인 17%를 상회했다고 가정) -> 약 922명 성공
variant_results = [922, 4349]

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 입력 폼 문구 및 버튼 디자인 변경이 전환율에 유의미한 상승을 이끌어냈습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 디자인 변경의 효과가 통계적으로 유의미하지 않습니다.")

#############
```

**결과해석**
